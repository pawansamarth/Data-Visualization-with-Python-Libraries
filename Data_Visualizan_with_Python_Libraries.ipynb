```python
## Q.10) Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap with seaborn.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Generate correlated data
np.random.seed(42)
n = 100
x1 = np.random.randn(n)
x2 = x1 + 0.5 * np.random.randn(n)
x3 = x1 - 0.3 * np.random.randn(n)
x4 = np.random.randn(n)

# Create a DataFrame
df = pd.DataFrame({'x1': x1, 'x2': x2, 'x3': x3, 'x4': x4})

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title('Correlation Matrix')
plt.show()
```

```python
## Q.13) Using the sales data, generate a heatmap to visualize the variation in sales across different months and days with plotly.
##       np.random.seed(20)
##       data = {'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
##                'Day': np.random.choice(range(1, 31), 100),  'Sales': np.random.randint(1000, 5000, 100)}
##       df = pd.DataFrame(data)
```

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)
data = {'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
        'Day': np.random.choice(range(1, 31), 100),
        'Sales': np.random.randint(1000, 5000, 100)}
df = pd.DataFrame(data)

# Create the heatmap
fig = px.imshow(df.pivot_table(index='Day', columns='Month', values='Sales'),
                 color_continuous_scale='Viridis',
                 title='Sales Variation Across Months and Days')
fig.show()
```

```python
!pip install plotly
import plotly.graph_objects as go
import plotly.express as px
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Provided data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
```

```python
# Create the 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Update layout for better visualization
fig.update_layout(
    title='3D Surface Plot of z = sin(√(x² + y²))',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis',
    ),
    width=800,
    height=800
)

# Show the plot
fig.show()
```

```python
## Q.17) Create a Bokeh scatter plot using randomly generated x and y values.
##       Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Generate random data
np.random.seed(42)
N = 100
x = np.random.rand(N) * 10
y = np.random.rand(N) * 10
sizes = np.random.randint(10, 50, N)
colors = np.random.rand(N, 3)  # Generate random RGB colors

# Create a ColumnDataSource
source = ColumnDataSource(data={'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a figure
p = figure(title="Scatter Plot with Varying Sizes and Colors")

# Add a scatter plot
p.scatter(x='x', y='y', size='sizes', color='colors', alpha=0.7, source=source)

# Show the plot
show(p)
```

```python
## Q.18) Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
##       fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']   counts = [20, 25, 30, 35]
from bokeh.plotting import figure, show

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a figure
p = figure(x_range=fruits, title="Fruit Counts")

# Add a bar glyph
p.vbar(x=fruits, top=counts, width=0.9)

# Show the plot
show(p)
```

```python
## Q.19) Create a Bokeh histogram to visualize the distribution of the given data.
##       data_hist = np.random.randn(1000)  hist, edges = np.histogram(data_hist, bins=30)
from bokeh.plotting import figure, show
import numpy as np

# Generate random data
np.random.seed(42)
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency')

# Add a bar glyph
p.vbar(x=edges[:-1], top=hist, width=edges[1] - edges[0])

# Show the plot
show(p)
```

```python
## Q.20) Create a Bokeh heatmap using the provided dataset.
##       data_heatmap = np.random.rand(10, 10)
##       x = np.linspace(0, 1, 10)
##       y = np.linspace(0, 1, 10)
##       xx, yy = np.meshgrid(x, y)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import linear_cmap

output_notebook()  # For displaying the plot in a Jupyter notebook
```

```python
# Provided data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
from bokeh.models import ColumnDataSource

# Flatten the data for Bokeh
source = ColumnDataSource(data=dict(
    x=xx.flatten(),
    y=yy.flatten(),
    value=data_heatmap.flatten(),
))
```

```python
# Define the color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Create the figure
p = figure(title="Heatmap Example", tools="hover", tooltips=[("Value", "@value")],
           x_axis_label="X-axis", y_axis_label="Y-axis",
           x_range=(0, 1), y_range=(0, 1))

# Add the heatmap
p.rect('x', 'y', width=0.1, height=0.1, source=source,
       fill_color=linear_cmap('value', 'Viridis256', data_heatmap.min(), data_heatmap.max()),
       line_color=None)

# Add the color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
